# Descubra as melhores notas de matemática do ENEM 2016

Você deverá criar um modelo para prever a nota da prova de matemática de quem participou do ENEM 2016.

O contexto do desafio gira em torno dos resultados do ENEM 2016 (disponíveis no arquivo train.csv). Este arquivo, e apenas ele, deve ser utilizado para todos os desafios. Qualquer dúvida a respeito das colunas, consulte o Dicionário dos Microdados do Enem 2016.

No arquivo test.csv crie um modelo para prever nota da prova de matemática (coluna **NU_NOTA_MT**) de quem participou do ENEM 2016.

Salve sua resposta em um arquivo chamado answer.csv com duas colunas: **NU_INSCRICAO** e **NU_NOTA_MT**.

Faça o upload do arquivo answer.csv usando o botão “Submeter resposta”.

Primeiro de tudo eu vou fazer a leitura dos arquivos. Para isso, eu upei uma cópia dos arquivos no meu github.

```
import pandas as pd

pd.options.display.float_format = '{:,.2f}'.format

uri = "https://github.com/hmalta/AceleraDev-DataScience/blob/master/train.csv?raw=true"
dados_train = pd.read_csv(uri)
dados_train.head()
```

```
uri = "https://github.com/hmalta/AceleraDev-DataScience/blob/master/test.csv?raw=true"
dados_test = pd.read_csv(uri)
dados_test.head()
```

Vou utilizar as mesmas colunas de notas que foram utilizadas nos exercícios da QuarentenaDev para treinar o modelo daqui.

```
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO', 'NU_NOTA_MT']
dados_train = dados_train[colunas_notas].dropna()
dados_train.head()
```

```
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
x_treino = dados_train[colunas_notas]
x_treino.head()
```

```
y_treino = dados_train['NU_NOTA_MT']
y_treino.head()
```

```
from sklearn.svm import LinearSVR

modelo = LinearSVR()
modelo.fit(x_treino, y_treino)
```

Até aqui, o modelo está treinado.

Agora vou separar os dados para predição e aplicar.

```
colunas_notas = ['NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO']
x_teste = dados_test[colunas_notas].dropna()
x_teste.head()
```

```
y_teste = modelo.predict(x_teste)
y_teste[:5]
```

Enfim, preciso juntar o y_teste à coluna de inscrição.

```
x_teste['NU_NOTA_MT'] = y_teste
x_teste.head()
```

```
dados_test_nu_inscr = dados_test['NU_INSCRICAO']
dados_test_nu_inscr.head()
```

```
x_teste = x_teste.join(dados_test_nu_inscr)
x_teste.head()
```

```
resultado = x_teste[['NU_INSCRICAO','NU_NOTA_MT']]
resultado
```

```
resultado.to_csv('answer.csv', index=False)
```
