# Descubra quem fez o ENEM 2016 apenas para treino

Neste desafio deverá descobrir quais estudantes estão fazendo a prova apenas para treino.

## Detalhes

O contexto do desafio gira em torno dos resultados do ENEM 2016 (disponíveis no arquivo train.csv). Este arquivo, e apenas ele, deve ser utilizado para todos os desafios. Qualquer dúvida a respeito das colunas, consulte o [Dicionário dos Microdados do Enem 2016](https://s3-us-west-1.amazonaws.com/acceleration-assets-highway/data-science/dicionario-de-dados.zip).

Alguns estudantes decidem realizar prova do ENEM de forma precoce, como um teste (coluna IN_TREINEIRO). Neste desafio, você deve criar um modelo de classificação binária para inferir a mesma. Os resultados possíveis da sua resposta devem ser “0” ou “1”.

Salve sua resposta em um arquivo chamado answer.csv com duas colunas: `NU_INSCRICAO` e `IN_TREINEIRO`.

```python
import pandas as pd
import numpy as np
import sklearn as sk
from sklearn.ensemble import RandomForestClassifier
```

```python
pd.options.display.float_format = '{:,.2f}'.format

dados_train = pd.read_csv("train.csv")
dados_train.head()
```

```python
dados_test = pd.read_csv("test.csv")
dados_test.head()
```

```python
features_test = list(dados_test.columns)
features_test.append('IN_TREINEIRO')
dados_train[features_test]
```

```python
corr = dados_train.corr()['IN_TREINEIRO'].sort_values(ascending=False)
corr.head(20)
```

```python
features = ['TP_ST_CONCLUSAO','TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC', 'TP_PRESENCA_MT']
dados_train[features]
```

```python
dados_train[features].isna().sum()
```

```python
x_treino = dados_train[features]
y_treino = dados_train['IN_TREINEIRO']
```

```python
clf = RandomForestClassifier(max_depth=2, random_state=0)
clf.fit(x_treino, y_treino)
```

```python
x_teste = dados_test[features]
```

```python
y_teste = clf.predict(x_teste)
```

```python
y_teste
```

```python
x_teste = x_teste.assign(IN_TREINEIRO = y_teste)
x_teste.head()
```

```python
dados_test_nu_inscr = dados_test['NU_INSCRICAO']
dados_test_nu_inscr.head()
```

```python
x_teste = x_teste.join(dados_test_nu_inscr)
x_teste.head()
```

```python
resultado = x_teste[['NU_INSCRICAO','IN_TREINEIRO']]
resultado
```

```python
resultado.to_csv('answer.csv', index=False)
```
